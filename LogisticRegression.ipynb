import necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

load file

```python
df = pd.read_csv('/content/drive/MyDrive/Student Stress Factors.csv')
```

```python
df.head()
```

```python
df = df.rename(columns={
    df.columns[1]: 'sleep_quality',
    df.columns[2]: 'headache',
    df.columns[3]: 'academic_performance',
    df.columns[4]: 'study_load',
    df.columns[5]: 'extra_activities',
    df.columns[6]: 'stress_levels'
})
```

```python
# Features: drop the target column
X = df.drop('stress_levels', axis=1)
X = df.drop('Timestamp', axis=1)

# Target: select the target column
Y = df['stress_levels']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42, stratify=Y)
print(f"Training set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")
```

```python
#Create a scalar object
scalar = StandardScaler()

# Fit and transform the training data
X_train_scaled = scalar.fit_transform(X_train)

# Transform the test data
X_test_scaled = scalar.transform(X_test)
```

```python
# Create and train model
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(f"Classification Report\n{classification_report(y_test, y_pred)}")
```
